```python
import pandas as pd
```

```python
gold_dt = pd.read_csv('./data/Gold Futures Historical Data.csv')
silver_dt = pd.read_csv('./data/Silver Futures Historical Data.csv')
platinum_dt = pd.read_csv('./data/Platinum Futures Historical Data.csv')
```

```python
gold_dt['Date'] = pd.to_datetime(gold_dt['Date'])
silver_dt['Date'] = pd.to_datetime(silver_dt['Date'])
platinum_dt['Date'] = pd.to_datetime(platinum_dt['Date'])
```

```python
r_idx_gold = [i for i in range(gold_dt.shape[0]-1,-1,-1)]
r_idx_silver = [i for i in range(silver_dt.shape[0]-1,-1,-1)]
r_idx_platinum = [i for i in range(platinum_dt.shape[0]-1,-1,-1)]
```

```python
gold_data = pd.DataFrame(gold_dt, index =r_idx_gold)
silver_data = pd.DataFrame(silver_dt, index =r_idx_silver)
platinum_data = pd.DataFrame(platinum_dt, index =r_idx_platinum)
```

```python
gold_data = gold_data.reset_index(drop=True)
silver_data = silver_data.reset_index(drop=True)
platinum_data = platinum_data.reset_index(drop=True)
```

```python
gold_df = gold_data[gold_data['Date'] > '2016-12-31']
gold_df2 = gold_df[gold_df['Date'] < '2022-01-01']
gold_df3 = gold_df2.reset_index()
gold = gold_df3.drop(['index'], axis=1)
gold
```

```python
silver_df = silver_data[silver_data['Date'] > '2016-12-31']
silver_df2 = silver_df[silver_df['Date'] < '2022-01-01']
silver_df3 = silver_df2.reset_index()
silver = silver_df3.drop(['index'], axis=1)
silver
```

```python
platinum_df = platinum_data[platinum_data['Date'] > '2016-12-31']
platinum_df2 = platinum_df[platinum_df['Date'] < '2022-01-01']
platinum_df3 = platinum_df2.reset_index()
platinum = platinum_df3.drop(['index'], axis=1)
platinum
```

```python
platinum.isna().sum()
```

```python
gold.to_csv('./data/gold.csv')
silver.to_csv('./data/silver.csv')
platinum.to_csv('./data/platinum.csv')
```
